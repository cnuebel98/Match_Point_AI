```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(test_df['C_Value'], [test_df['N_Simus'], test_df['Point_Winner']])

# Perform the Chi-squared test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the results
print(f"Chi-squared value: {chi2}")
print(f"P-value: {p}")

# Compare the p-value to a significance level (e.g., 0.05) to make a decision
if p < 0.05:
    print("There is a significant relationship between Point_Winner and at least one of the other two columns.")
else:
    print("There is no significant relationship between Point_Winner and the other two columns.")
```
