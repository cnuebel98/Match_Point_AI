```python
import pandas as pd
from scipy.stats import chi2_contingency

# Assuming df_1 is your DataFrame with 'Point_Winner' and 'Sel_Pol' columns
# Example data
df_1 = test_df

# Create a contingency table
contingency_table = pd.crosstab(df_1['Point_Winner'], df_1['Sel_Pol'])

# Perform the Chi-Square test
chi2, p, _, _ = chi2_contingency(contingency_table)

# Print the results
print("Chi-Square Value:", chi2)
print("P-value:", p)

# Interpret the results
alpha = 0.05
print("\nSignificance Level (alpha):", alpha)
print("Conclusion:")
if p < alpha:
    print("Reject the null hypothesis. There is a significant association between Point_Winner and Sel_Pol.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between Point_Winner and Sel_Pol.")

print("Contingency Table:")
print(contingency_table)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

df_1 = test_df

# Create a contingency table with three categorical variables
contingency_table = pd.crosstab(index=[df_1['Sel_Pol'], df_1['Dec_Pol'], df_1['Opponent']], columns=df_1['Point_Winner'])

# Perform the Chi-Square test
chi2, p, _, _ = chi2_contingency(contingency_table)

# Print the results
print("Chi-Square Value:", chi2)
print("P-value:", p)

# Interpret the results
alpha = 0.05
print("\nSignificance Level (alpha):", alpha)
print("Conclusion:")
if p < alpha:
    print("Reject the null hypothesis. There is a significant association among the variables.")
else:
    print("Fail to reject the null hypothesis. There is no significant association among the variables.")

print("Contingency Table:")
print(contingency_table)

from statsmodels.stats.contingency_tables import Table

# Create a contingency table with statsmodels
contingency_table_statsmodels = Table(contingency_table)

# Calculate residuals
residuals = contingency_table_statsmodels.resid_pearson

# Display residuals
print("Residuals:")
print(residuals)
```
