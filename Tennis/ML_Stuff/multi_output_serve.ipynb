# Here we have a model that predicts the 1st Serve of Novak djokovic (dir & specs) with an accuracy of 20,19%

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\carlo\TrainingsTool\Tennis\ML_Stuff\first_serve_pred_djoko_data.csv", encoding="unicode_escape", low_memory=False)
df
```

```python
df = df.replace(np.nan, '', regex=True)
df
```

```python
from sklearn.datasets import make_multilabel_classification
from sklearn.svm import SVC
```

```python
df.drop("Unnamed: 0", axis = 1, inplace=True)
df
```

```python
X = df.drop(["0_dir", "0_spec"], axis=1)
X
```

```python
y = df[["0_dir", "0_spec"]]
y
```

```python
clf = MultiOutputClassifier(RandomForestClassifier())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
clf.fit(X_train, y_train)
```

```python
y_preds = clf.predict(X_test)
y_preds
```

```python
y_test
```

```python
clf.score(X_test, y_test)
```

```python
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators...")
    clf = MultiOutputClassifier(RandomForestClassifier(n_estimators=i)).fit(X_train, y_train)
    print(f"Model accuracy on test set: {clf.score(X_test, y_test) * 100:.2f}%")
    print("")
```
